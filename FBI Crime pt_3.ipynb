# Regression Models 

### small cities 

```python
X = small_cities[["Burglary_sqrt","Sqr_Pop","Larceny_theft"]]
y = small_cities["Property_crime"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
regr_small = linear_model.LinearRegression()
regr_small.fit(X_train, y_train)
regr_small.score(X_train, y_train)
```

```python
# Extract predicted values.
predicted = regr_small.predict(X_train).ravel()
actual = y_train

# Calculate the error, also called the residual.
residual = actual - predicted

# This looks a bit concerning.
plt.hist(residual)
plt.title('Residual counts')
plt.xlabel('Residual')
plt.ylabel('Count')
plt.show()
```

```python
regr_small.score(X_test, y_test)
```

```python
# Extract predicted values.
predicted = regr_small.predict(X_test).ravel()
actual = y_test

# Calculate the error, also called the residual.
residual = actual - predicted

# This looks a bit concerning.
plt.hist(residual)
plt.title('Residual counts')
plt.xlabel('Residual')
plt.ylabel('Count')
plt.show()
```

### cross validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(regr_small, X_train, y_train, cv=10)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(regr_small, X_test, y_test, cv=10)
```
